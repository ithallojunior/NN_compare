```python
#this snippet of code gets the data from .mat files into the running program
#importing the libraries

import numpy as np
```

```python
#this function considers each column to be a different input 
def read_data(archive, rows, columns):
    data = open(archive, 'r')
    mylist = data.read().split()
    data.close()
    myarray = np.array(mylist).reshape((columns, rows)).astype(float)
    return myarray
    

            
```
